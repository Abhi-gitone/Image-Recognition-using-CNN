**Dataset**: Standard cat-dog stuff.

**Task**: Image recognition using Convolutional Neural Network (CNN)

```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

```python
tf.__version__
```

```python
## ImageDataGenerator --> used to load, augment, and scale image data
```

```python
## Train set

## Image augmentation --> geometrical transformations on train data to prevent overfitting (flip, zoom, rotate etc.)

train_generator = ImageDataGenerator(
        rescale=1./255, ## scale image vector so all values lie betn 0 and 1 for NN
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)

## load train data using generator

train = train_generator.flow_from_directory(
        'dataset/training_set',
        target_size=(64, 64), ## desired image pixel size
        batch_size=32,
        class_mode='binary')
```

```python
## Test set

## Test data is never augmented

test_generator = ImageDataGenerator(rescale=1./255)

test = test_generator.flow_from_directory(
        'dataset/test_set',
        target_size=(64, 64),
        batch_size=32,
        class_mode='binary')
```

```python
type(train)
```

```python
type(test)
```

```python
## train and test are generator objects --> save memory
```

```python
## instantiation
cnn = tf.keras.models.Sequential()

## convolution layer 1
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu', input_shape=(64,64,3)))
## 32 feature detectors, feature detector of size 3X3, 'relu' adds non-linearity, 3D vector for RGB image of size 64X64

## pooling layer 1
cnn.add(tf.keras.layers.MaxPool2D(pool_size=(2,2), strides=2))

## convolution layer 2
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu'))
## 32 feature detectors, feature detector of size 3X3, 'relu' adds non-linearity

## pooling layer 2
cnn.add(tf.keras.layers.MaxPool2D(pool_size=(2,2), strides=2))

## flattening
cnn.add(tf.keras.layers.Flatten())

## full connection 1
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
## dropout layer to prevent overfitting
cnn.add(tf.keras.layers.Dropout(rate=0.2)) ## randomly turns off 20% neurons at every iteration

## full connection 2
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
## dropout layer to prevent overfitting
cnn.add(tf.keras.layers.Dropout(rate=0.2)) ## randomly turns off 20% neurons at every iteration

## output layer
cnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
## 'sigmoid' --> binary classification, 'softmax' --> multi-class classification

## compilation
cnn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
cnn.summary()
```

```python
## instantiating callback

tensorboard_callback = tf.keras.callbacks.TensorBoard()
```

```python
## model fit

history = cnn.fit(train, validation_data=test, epochs=25, callbacks=[tensorboard_callback])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## accuaracy over epochs

fig, ax = plt.subplots(figsize=(16,6))

ax.plot(history.history['acc'])
ax.plot(history.history['val_acc'])
ax.set_xlabel('Epochs')
ax.set_ylabel('Accuracy')
ax.legend(['Train', 'Test'])

plt.show()
```

```python
## loss over epochs

fig, ax = plt.subplots(figsize=(16,6))

ax.plot(history.history['loss'])
ax.plot(history.history['val_loss'])
ax.set_xlabel('Epochs')
ax.set_ylabel('Loss')
ax.legend(['Train', 'Test'])

plt.show()
```

```python
## creating a function to predict

def predict(model, image_location):
    
    img = image.load_img(image_location, target_size=(64,64))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    result = model.predict(img/255.)
    
    if result[0][0] >= 0.5:
        print("dog")
    else:
        print("cat")
```

```python
predict(model=cnn, image_location='dataset/single_prediction/cat_or_dog_1.jpg')
```
